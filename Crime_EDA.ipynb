```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('crime_data.csv')
```

```python
print(df.head())
```

```python
plt.figure(figsize=(18,5))
df['City'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Crime per City')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Victim Age'], bins=10, kde=True, color='purple')
plt.title('Distribution of Victim Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.boxplot(data=df, x='Crime Domain', y='Police Deployed', palette='Set2', legend=False, hue='Crime Domain')
plt.title('Police Deployment by Crime Domain')
plt.xlabel('Crime Domain')
plt.ylabel('Number of Police Deployed')
plt.tight_layout()
plt.show()
```

```python

df['Hour'] = pd.to_datetime(df['Time of Occurrence'], format='%d-%m-%Y %H:%M').dt.hour

# Plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Hour'], bins=24, kde=False, color='orange')
plt.title('Distribution of Crimes by Hour of Day')
plt.xlabel('Hour (0-23)')
plt.ylabel('Number of Crimes')

plt.tight_layout()
plt.show()
```

```python
sns.countplot(y='Crime Description', data=df, order=df['Crime Description'].value_counts().index)
```

```python
sns.countplot(y='Weapon Used', data=df, order=df['Weapon Used'].value_counts().index)
```

```python
sns.boxplot(x='Weapon Used',data=df,  y=df['Hour'])
```

```python
sns.barplot(x='Crime Description', y='Case Closed', data=df,  estimator='mean')
```

```python
sns.heatmap(pd.crosstab(df['City'], df['Crime Description']), annot=True, cmap='Blues', fmt='d')
plt.show()
```
